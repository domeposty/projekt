```python
import os
import numpy as np
import cv2
import tqdm
import random
import pickle
```

```python
classifier = cv2.CascadeClassifier(cv2.data.haarcascades+"haarcascade_frontalface_default.xml")
videosnumber = next(os.walk('StressVideos'))[2]
parent_dir = 'StressVideos'
IMG_SIZE=50
current_video=1
current_frame=1
for i in range(len(videosnumber)):
    directory = 'video_' + str(current_video) + '_faces'
    dir_path = os.path.join(parent_dir, directory)
    new_dir = os.mkdir(dir_path)
    print(f'Creating: {directory}')
    video_name = 'video_' + str(current_video) + '.mp4'
    video_path=os.path.join(parent_dir, video_name)
    cap = cv2.VideoCapture(video_path)
    path_to_save = dir_path
    if (cap.isOpened()== False):
        print('Cap is not open')
    while(cap.isOpened()):
        ret, frame = cap.read()
        if(ret==True):
            faces = classifier.detectMultiScale(
            frame,  
            scaleFactor=1.05,
            minNeighbors=20, 
            minSize=(100, 100) 
            ) 
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x + w, y + h),(0,0,255),thickness=2)
                sub_face = frame[y:y+h, x:x+w]
            name = 'face_' + str(current_frame) + '.jpg'
            #print(f'Creating: {name}')
            cv2.imwrite(os.path.join(path_to_save, name), sub_face)
            current_frame +=1
        else:
            break
    cap.release()
    print('done')
    FacesData= []
    def create_FacesData(): 
        path = os.path.join(dir_path)
        for img in os.listdir(dir_path): 
            img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
            new_array=cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) 
            FacesData.append([new_array]) 
    create_FacesData()
    random.shuffle(FacesData)
    Faces=[]
    for features in FacesData:
        Faces.append(features)
    Faces=np.array(Faces).reshape(-1,IMG_SIZE, IMG_SIZE, 1)
    Faces=Faces/255
    pickle_out = open("Faces_" + str(current_video) + ".pickle", "wb")
    pickle.dump(Faces, pickle_out)
    pickle_out.close()
    current_video +=1 
print('done')
videosnumber = next(os.walk('NonstressVideos'))[2]
parent_dir = 'NonstressVideos'
current_video=1
current_frame=1
for i in range(len(videosnumber)):
    directory = 'video_' + str(current_video) + '_faces'
    dir_path = os.path.join(parent_dir, directory)
    new_dir = os.mkdir(dir_path)
    print(f'Creating: {directory}')
    video_name = 'video_' + str(current_video) + '.mp4'
    video_path=os.path.join(parent_dir, video_name)
    cap = cv2.VideoCapture(video_path)
    path_to_save = dir_path
    if (cap.isOpened()== False):
        print('Cap is not open')
    while(cap.isOpened()):
        ret, frame = cap.read()
        if(ret==True):
            faces = classifier.detectMultiScale(
            frame,  
            scaleFactor=1.05,
            minNeighbors=20, 
            minSize=(100, 100) 
            ) 
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x + w, y + h),(0,0,255),thickness=2)
                sub_face = frame[y:y+h, x:x+w]
            name = 'face_' + str(current_frame) + '.jpg'
            #print(f'Creating: {name}')
            cv2.imwrite(os.path.join(path_to_save, name), sub_face)
            current_frame +=1
        else:
            break
    cap.release()
    print('done')
    FacesData= []
    def create_FacesData(): 
        path = os.path.join(dir_path)
        for img in os.listdir(dir_path): 
            img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
            new_array=cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) 
            FacesData.append([new_array]) 
    create_FacesData()
    random.shuffle(FacesData)
    Faces=[]
    for features in FacesData:
        Faces.append(features)
    Faces=np.array(Faces).reshape(-1,IMG_SIZE, IMG_SIZE, 1)
    Faces=Faces/255
    pickle_out = open("FacesN_" + str(current_video) + ".pickle", "wb")
    pickle.dump(Faces, pickle_out)
    pickle_out.close()
    current_video +=1 
print('done')
```
