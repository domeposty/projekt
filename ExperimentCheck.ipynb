```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd        
```

```python
Results=[3,3,9,4,7,10,5,2,1,6,8, 5, 6, 4, 1, 3, 5, 2,6,6]  
bin_count = int(np.ceil(np.log2(len(Results)))+1) #Sturge rule for bin count
res = pd.Series(Results, name="Broj Tocnih Odgovora")
sns.displot(stat="count",data=res, kind="hist", linewidth=0.5, kde=True, bins=20, color=plt.style.use('dark_background'))
sns.displot(stat="frequency",data=res, kind="hist", linewidth=0.5, kde=True, bins=20,color=plt.style.use('dark_background'))
sns.displot(stat="probability",data=res, kind="hist", linewidth=0.5, kde=True, bins=20,color=plt.style.use('dark_background'))
sns.displot(stat="count",data=res, kind="hist", linewidth=0.5, kde=True, bins=bin_count,color=plt.style.use('dark_background'))
```
