```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import os 
```

```python
current_video = 1
save_path = input ("enter the save path : ")
folder_path = input ("enter the folder path : ")
file_extn = input ("enter the video file extension like mp4 or mov or avi : ")
complete_video_file = []
for video_file in os.listdir(folder_path):
    if video_file.endswith(file_extn):
        complete_video_file.append(os.path.join(folder_path,video_file))
for required_video_file in complete_video_file:  
    with open("times.txt") as f:
        times = f.readlines()
    times = [x.strip() for x in times] 
    for time in times:
        starttime = int(time.split('-')[0])
        endtime = int(time.split("-")[1])
        ffmpeg_extract_subclip(required_video_file, starttime, endtime, targetname=os.path.join(save_path, 'video_' + str(current_video) + ".mp4"))
        current_video +=1
```
